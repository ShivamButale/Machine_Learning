```python
from scipy import stats
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
# array x contains size of houses
x = np.array([112,345,198,306,372,660,302,420,578]) 
```

```python
#array y  coontains corresponding prices
y = np.array([1120,1523,2102,2230,2600,3200,3409,3689,4460])
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

```

```python
plt.plot(x,y,'ro', color='black')

```

```python
plt.ylabel('Price')
plt.xlabel('Size of Houses')
```

```python
plt.axis([0,600,0,5000])
```

```python
plt.plot(x,x*slope+intercept, 'b')
```

```python
plt.plot()
plt.show()
```
